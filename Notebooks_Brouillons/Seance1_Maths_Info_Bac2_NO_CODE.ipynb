## 1) Logique de base
### 1.1 Vocabulaire
- **Proposition** : phrase qui est *vraie* ou *fausse* (ex. *“2 est pair”*).  
- **Variables** : \(p, q, r\) représentent des propositions.  
- **Connecteurs** :
  - **NON** : \(\neg p\) (ou *non p*)  
  - **ET** : \(p \land q\)  
  - **OU** : \(p \lor q\) (*au moins un vrai*)  
  - **IMPLIQUE** : \(p \rightarrow q\) (*si \(p\) alors \(q\)*)  
  - **ÉQUIVALENT** : \(p \leftrightarrow q\)
- **Priorité** (comme en calcul) : \(\neg\) > \(\land\) > \(\lor\) > \(\rightarrow\) > \(\leftrightarrow\).  
  → **Toujours** mettre des parenthèses si doute.

### 2.1 Table pour \(p\) et \(q\)

### 2.2 Implication & équivalence
| \(p\) | \(q\) | \(p \rightarrow q\) | \(p \leftrightarrow q\) |
|---|---|---|---|
| 0 | 0 |   |   |
| 0 | 1 |   |   |
| 1 | 0 |   |   |
| 1 | 1 |   |   |

<details><summary><b>Correction guidée</b> (cliquer)</summary>

- \(p \rightarrow q\) est **fausse seulement** quand \(p=1\) et \(q=0\).  
- \(p \leftrightarrow q\) est **vraie** quand \(p\) et \(q\) ont la **même** valeur.

</details>

## 3) Équivalences utiles (à connaître)
1. **Implication** : \(p \rightarrow q \equiv \neg p \vee q\)  
2. **De Morgan** : \(\neg(p\wedge q) \equiv \neg p \vee \neg q\) ;  \(\neg(p\vee q) \equiv \neg p \wedge \neg q\)  
3. **Biconditionnel** : \(p \leftrightarrow q \equiv (p\wedge q)\vee(\neg p \wedge \neg q)\)

### 3.1 Vrai ou faux ? (justifiez avec une table ou un exemple)
a) \(\neg(p \vee q) \equiv \neg p \vee \neg q\)  
b) \(p \rightarrow q \equiv \neg q \rightarrow \neg p\) (contraposée)
  
<details><summary><b>Indications</b></summary>
a) Attention : c’est **faux** (au lieu de OU, c’est **ET**).  
b) **Vrai** : la contraposée a toujours la même valeur de vérité que l’implication.
</details>

## 5) Ensembles — dédoublonnage (idée sans code)
- Un **ensemble** ne contient **pas de doublon**.  
- Si une liste contient un doublon, alors \(|\text{ensemble}(\text{liste})| < |\text{liste}|\).  
- Exemple : \(["a","b","a","c"]\) → ensemble = \(\{a,b,c\}\) (taille 3 < 4).
